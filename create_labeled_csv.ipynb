```python

# rename files in a directory or folder 
  
# importing os module 
import os 
  
# Function to rename multiple files according to their folder
def main(): 
    
    main_folders = ['apple','banana','cucumber','tomato','lemon','orange']
   
    sub_folders = ['Apples','Banana','Cucumber','tomato','Lemon','Orange']
    
    train_test = ['train','test']
    fresh_rotten = ['fresh','medium','rotten']
    
    for ind, f in enumerate(main_folders):
        path = "E:\\FinalDataset\\"+f+"\\"+sub_folders[ind]
        for i in range(2):
            for j in range(3):
                new_path = path+"\\"+train_test[i]+"\\"+fresh_rotten[j]+"\\"
                for count, filename in enumerate(os.listdir(new_path)):
                    dst = main_folders[ind]+"_"+train_test[i]+"_"+fresh_rotten[j] +"_"+ str(count+1) + ".jpg"
                    src = new_path+ filename 
                    dst = new_path+ dst 

                    # rename() function will 
                    # rename all the files 
                    os.rename(src, dst) 
      
    
  
# Driver Code 
if __name__ == '__main__': 
      
    # Calling main() function 
    main() 
```

```python
import pandas as pd
# Function to create .csv file with filenames and labels
def main(): 
    
    main_folders = ['apple','banana','cucumber','tomato','lemon','orange']
    sub_folders = ['Apples','Banana','Cucumber','tomato','Lemon','Orange']

    sub_folders = ['Cucumber','tomato','Orange']
    train_test = ['train','test']
    fresh_rotten = ['fresh','medium','rotten']
    
    for ind, f in enumerate(main_folders):
        path = "E:\\FinalDataset\\"+f+"\\"+sub_folders[ind]
        files = []
        labels = []
        for i in range(2):
            for j in range(3):
                new_path = path+"\\"+train_test[i]+"\\"+fresh_rotten[j]+"\\"
                for count, filename in enumerate(os.listdir(new_path)):
                    files.append(filename)
                    labels.append(fresh_rotten[j])
        
#         df = pd.DataFrame(files,labels, columns=['name','label'])
        df = pd.DataFrame()
        df["name"]=files
        df["label"]=labels
        df.to_csv(path+"3.csv")
      
    
  
# Driver Code 
if __name__ == '__main__': 
      
    # Calling main() function 
    main() 
```
