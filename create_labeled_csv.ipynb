```python

# rename files in a directory or folder 
  
# importing os module 
import os 
  
# Function to rename multiple files according to their folder
def main(): 
    
    #main folders or directories
    main_folders = ['apple','banana','cucumber','tomato','lemon','orange']
   
    #sub_folders or directories inside main folders
    sub_folders = ['Apples','Banana','Cucumber','tomato','Lemon','Orange']
    
    #sub sub folders
    train_test = ['train','test']
    #sub sub sub folders
    classes = ['fresh','medium','rotten']
    
    for ind, f in enumerate(main_folders):
        #base directory path
        path = "E:\\FinalDataset\\"+f+"\\"+sub_folders[ind]
        #iterate on sub sub folders
        for i in range(2):
            #iterate on sub sub sub folders
            for j in range(3):
                #path to the sub sub sub folder
                new_path = path+"\\"+train_test[i]+"\\"+classes[j]+"\\"
                #iterate on all the images in the new_path
                for count, filename in enumerate(os.listdir(new_path)):
                    #assigning name to image as (main_folder +_+ sub_folder +_+ sub_sub_sub_folder +_+ count).jpg
                    dst = main_folders[ind]+"_"+train_test[i]+"_"+classes[j] +"_"+ str(count+1) + ".jpg"
                    #complete path of source image
                    src = new_path+ filename
                    #complete path of new modified named image
                    dst = new_path+ dst 

                    # rename() function will 
                    # rename all the files 
                    os.rename(src, dst) 
      
    
  
# Driver Code 
if __name__ == '__main__': 
      
    # Calling main() function 
    main() 
```

```python
import pandas as pd
# Function to create .csv file with filenames and labels
def main(): 
    
    main_folders = ['apple','banana','cucumber','tomato','lemon','orange']
    sub_folders = ['Apples','Banana','Cucumber','tomato','Lemon','Orange']

    train_test = ['train','test']
    classes = ['fresh','medium','rotten']
    
    for ind, f in enumerate(main_folders):
        #base directory path
        path = "E:\\FinalDataset\\"+f+"\\"+sub_folders[ind]
        #files will hold names of all images while labels will hold their corresponding labels like fresh,rotten,medium 
        files = []
        labels = []
        #iterate on sub sub folders
        for i in range(2):
            #iterate on sub sub sub folders
            for j in range(3):
                #path to the sub sub sub folder
                new_path = path+"\\"+train_test[i]+"\\"+classes[j]+"\\"
                #iterate on all the images in the new_path
                for count, filename in enumerate(os.listdir(new_path)):
                    files.append(filename)
                    labels.append(classes[j])
        
        
        df = pd.DataFrame()
        df["name"]=files
        df["label"]=labels
        df.to_csv(path+"3.csv")
      
    
  
# Driver Code 
if __name__ == '__main__': 
      
    # Calling main() function 
    main() 
```
